```python
import pyspark.sql.functions as F
from pyspark.sql import DataFrame as SparkDataFrame
```

```python
routes: SparkDataFrame = catalog.load("routes@spark")
```

```python
routes.show(truncate=False)
```

```python
routes.printSchema()
```

```python
route_df = routes\
    .withColumn("data_exploded", F.explode(F.col("data")))\
    .withColumn("file_name", F.input_file_name())\
    .select(
        F.col("data_exploded")
    )
```

```python
route_df.show(truncate=False)
```

```python
route_df.printSchema()
```

```python
route_df.select(
    F.col("data_exploded.id"),
    F.col("data_exploded.attributes"),
    F.col("data_exploded.attributes.description"),
    F.col("data_exploded.attributes.direction_destinations"),
    F.col("data_exploded.attributes.direction_names"),
    F.col("data_exploded.attributes.fare_class"),
    F.col("data_exploded.attributes.listed_route"),
    F.col("data_exploded.attributes.long_name"),
    F.col("data_exploded.attributes.short_name"),
    F.col("data_exploded.relationships.agency.data.id").alias("agency_id"),
    F.col("data_exploded.relationships.line.data.id").alias("line_id"),
).show(truncate=False)
```
